```python
#let's create a function to create adjusted R-squared
def adj_r2(x,y):
    r2=lr.score(x,y)
    n=x.shape[0] #no. of rows
    p=x.shape[1] #no. of cols
    adjusted_r2=1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```

```python
adj_r2(x_test,y_test)
#adjusted R-square is little bit lesser than R-square
#this is our actual accuracy(may be)
```

```python
#Implementation of Lasso:

#Lasso cross validation
#Cross validation is a technique to find the best possible parameter by doing random exp
#cv will be used heavily as hyper parameter
#here alphaas is same as lambda
#If cv=5, then it will divide our dataset(whether train or test or whole dataset) into 5 sets
lassocv=LassoCV(alphas=None, cv=5, max_iter=2000000, normalize=True)
lassocv.fit(x_train, y_train)
#initially alphas=None
```

```python
#optimal value of lambda. Lambda is unknown variable
lassocv.alpha_
```

```python
lasso=Lasso(alpha=lassocv.alpha_)
lasso.fit(x_train, y_train)
```

```python
lasso.score(x_test, y_test)
```

```python
#This is a linear regression model with Lasso

#If CV=5, then it wil divide our data into 5 sets and 4 sets will be used for training and 1 for testing;for 2nd iteration again it wil choose another 4 for traing and 1 for testing; and everytime it wil keep rotating the values of 4 and 1 dataset. If CV=50, then it wil divide our data into 50 sets and 49 sets will be used for training and 1 for testing.
```

```python
#EVEN IN RIDGE, WE ARE SUPPOSED TO FIND OUT BEST POSSIBLE ALPHA VALUES

ridgecv=RidgeCV(alphas=(0.1, 1.0, 10.0), cv=10, normalize=True)
ridgecv.fit(x_train, y_train)
#argument used in alphas is for the length of size
```

```python
#out of 3 alphas, selwct 1 possible one
ridgecv.alpha_
```

```python
#we will try to keep higher range of alpha bcoz its a hyperparameter
np.random.uniform(0,10,50)
#50 records b/w 0 to 10
#we will give this data now in alphas parameter
```

```python
ridgecv=RidgeCV(alphas=np.random.uniform(0,10,50), cv=10, normalize=True)
ridgecv.fit(x_train, y_train)
```

```python
ridgecv.alpha_
#now out of 50 datasets that we have given, it is able to finalizealpha value as this one
```

```python
#we will create ridge object with alpha
ridge_lr=Ridge(alpha=ridgecv.alpha_)
ridge_lr.fit(x_train, y_train)
```

```python
#score of model with testing
ridge_lr.score(x_test, y_test)
```
